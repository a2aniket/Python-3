```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection  import train_test_split
from keras.layers import Dense,Activation
from keras.models import Sequential
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
data=pd.read_csv("BankCustomers.csv")
data
```

```python
numerical_feature=[feature for feature in data.columns if data[feature].dtype!="O"]
categorical_feature=[feature for feature in data.columns if data[feature].dtype=="O"]
```

```python
geography=pd.get_dummies(data["Geography"],prefix=True,drop_first=True)
gender=pd.get_dummies(data["Gender"],prefix=True,drop_first=True)
```

```python
data=pd.concat([data,gender,geography],axis=1)
data=data.drop(["Geography","Gender"],axis=True)
data
```

```python
x=data.iloc[:,3:]
x.drop(["Exited"],axis=1,inplace=True)
y=data["Exited"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
scaller=MinMaxScaler()
x_train=scaller.fit_transform(x_train)
```

```python
def model_building(layers,activation,kernel_initializer):
    model=Sequential()
    for i,node in enumerate(layers):
        if i!=0:
            model.add(Dense(node))
            model.add(Activation(activation))
    else:
            model.add(Dense(node,input_dim=1))
            model.add(Activation(activation))
            
    model.add(Dense(1,activation="sigmoid"))
    model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
    return model
```

```python
model=KerasClassifier(model_building,verbose=1)
param=GridSearchCV(estimator=model,param_grid=grid_param)
param_result=param.fit(x_train,y_train,verbose=1)
```
